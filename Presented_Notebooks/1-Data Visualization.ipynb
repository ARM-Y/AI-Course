### 4-Bar Charts
Session 4 / slide 19 bar plot can be seen:

```python
# Our DataFrame is named df and the column we want to plot is 'children'
name_of_column = df['children']

# Getting unique categories in the column
categories = name_of_column.unique()

#Getting the count of each category
category_counts = name_of_column.value_counts()

# Set a custom palette. We can change the palettes and the colors of our plot will change
custom_palette = sns.color_palette('Set2')  # Example: using the 'Set2' palette from seaborn

# Create a bar plot
plt.bar(categories, category_counts, color=custom_palette)

# Set the label for the x-axis
plt.xlabel('Number of Children')

# Set the label for the y-axis
plt.ylabel('Counts')

# Set the title of the bar plot
plt.title('Bar Plot')

# Display the bar plot
plt.show()
```

### 5-Pie Charts

```python
# Our DataFrame is named df and the column we want to plot is 'smoker'
name_of_column = df['smoker']

# Getting unique categories in the column
categories = name_of_column.unique()

#Getting the count of each category
category_counts = name_of_column.value_counts()

# Create a pie chart
plt.pie(category_counts, labels=categories)

# Create a pie chart with percentage labels
plt.pie(category_counts, labels=categories, autopct='%1.1f%%')

# Set the title of the pie chart
plt.title('Pie Chart of Smokers and non Smokers')

# Display the pie chart
plt.show()
```
